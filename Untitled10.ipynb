```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from matplotlib.animation import FuncAnimation
from sklearn.datasets import load_boston
from sklearn.preprocessing import MinMaxScaler
from IPython.display import HTML
from sklearn.metrics import mean_squared_error
```

```python
boston = load_boston()
```

```python
features = pd.DataFrame(boston.data, columns=boston.feature_names)
```

```python
features
```

```python
def update(m,x,c,t,learning_rate):
    grad_m=sum((2*(m * x + c)- t) * x)
    grad_c=sum((2*(m * x + c)- t))
    m = m - grad_m * learning_rate
    c = c - grad_c * learning_rate
    return m,c
```

```python
target = pd.DataFrame(boston.target, columns=['target'])
```

```python
target
```

```python
df = pd.concat([features, target], axis=1)
df
```

```python
x = df['LSTAT'].values
y = df['target'].values
print(y[:5])
```

```python
x_scaler = MinMaxScaler()
x = x_scaler.fit_transform(x.reshape(-1, 1))
x = x[:, -1]
y_scaler = MinMaxScaler()
y = y_scaler.fit_transform(y.reshape(-1, 1))
y = y[:, -1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=6)
```

```python
def error(m,x,c,t):
    N = x.size
    e = sum(((m*x+c)-t)**2)
    return e*1/(2*N)
```

```python
def gradient_descent(init_m, init_c, x, t, learning_rate, iterations, error_threshold):
  m = init_m
  c = init_c
  error_values = list()
  mc_values = list()
  for i in range(iterations):
    e = error(m, x, c, t)
    if e<error_threshold:
      print('error less than the threshold. Stopping gradient function')
      break
    error_values.append(e)
    m, c = update(m,x,c,t,learning_rate)
    mc_values.append((m,c))
  return m,c,error_values, mc_values
```

```python
%%time
init_m=0.9
init_c=0
learning_rate=0.001
iterations=250
error_thershold=0.001
m,c,error_values,mc_values=gradient_descent(init_m,init_c,x_train,y_train,learning_rate,iterations,error_thershold)
```

```python
#as rhe number of iterations increases,change in lines are less noticable
#Inorder to reduce the processing time in animation,it is advised to use smaller values
mc_values_anim = mc_values[0:250:5]


```

```python
plt.scatter(x_train, y_train, color='b')
plt.plot(x_train, (m*x_train+c), color='r')
```

```python
plt.plot(np.arange(len(error_values)), error_values)
plt.ylabel('Error')
plt.xlabel('Iterations')
```

```python
fig, ax = plt.subplots()
ln, = plt.plot([], [], 'ro-', animated=True)
def init():
  plt.scatter(x_test, y_test, color='g')
  ax.set_xlim(0, 1.0)
  ax.set_ylim(0, 1.0)
  return ln,

def update_frame(frame):
  m, c = mc_values_anim[frame]
  x1, y1 = -0.5, m*-0.5 + c
  x2, y2 = 1.5, m*1.5+c
  ln.set_data([x1, x2], [y1, y2])
  return ln,

anim = FuncAnimation(fig, update_frame, frames=range(len(mc_values_anim)),init_func=init, blit=True)
HTML(anim.to_html5_video())  
```
